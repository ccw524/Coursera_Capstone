##  Comparing New York to Toronto

Introduction/Business Problem:

New York and Toronto are both the largest cities, population wise, in the U.S and Canada. For my final assignment, I decided to look at the cities as a whole to see how the clustering in the two cities compare. While the physical size and the population of Toronto is smaller than New York, it is still interesting to see how the two compare.

The reason why I focused on this, was based on a discussion at work. The company that I work for has a vested business interest and presence in both cities. Some of my co-workers, who have travelled to both cities, were talking about their experiences there as they had stayed to explore more of the city. Recently I had the opportunity to New York for business and will be going several time in the coming year with a possbility of having a trip to Toronto, looking into what venues the most common for different areas.

Data:

For this assignment, I used the same datasets as the New York and Toronto assignments. However, instead of focusing on just one borough in New York or just neighborhoods in Toronto that contained "Toronto", I looked at all of the neighborhoods for both cities. While both cities are the largest in their respective countries, in addition to covering more square miles, New York's larger population allows for more venues in the area. However performing several cluster analysis with different changes for the inputs the results were interesting.

Methodology & Results:

In the original assignments, we looked at a small part of Toronto (Neighborhoods that contiained the name Toronto) and then Manhattan, which is thde smallest of the five boroughs in New York. Since New York is the bigger city, I spent the majority of my time going through the different boroughs and looking at the clusters for each borough to compare to Manhattan.

One of my initial idea I had was to compare one other borough's clusters against the clusers that I developed for the assignment. After looking at the cluster for Brooklyn and Queens, I decided to look at how the clusters would look like when I used the whole city. In the Toronto assignment, I specified to only include Boroughs that contained "Toronto" in its name. Similar to New York, I decided to focus on the entire city as I am not as familar with Toronto or it's boroughs. 

Once I ran the initial clustering for both cities, I went through and looked at how the numbers within cluster shifted by just adjusting the number of points fell within each cluster. My default was 5 clusters, like in the previous assingments. As the numbers of clusters increased, I noticed that for both cities the number in each cluster would dramaticly shift with most being in one or two clusters. This lead me to keep my clusters at 5.

Next I consulted the KMeans information site, https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html , to see which of the parametrics for the function to run I can tune to see how the clusters would be divided. The first parameter that I changed was "init", which changes where the initization of the clusters begins for each start. I changed the default value of "k-means++" to "random", which will "choose k observations (rows) at random from data for the initial centroids". Next, I ran the clustering model again with various clusters to see if the initialization of the clusters to see what changed. 

For New York, the first cluster, which primarily was in Staten Island, had a most common venue of Bus Stop's. After that the 2nd most common venue varies, but with almost half being a public location. The second cluster, which had the largest points, most common venue was almost split between Pizza Places and Italian restaurants. The second most common venue was also pizza place but with Deli/Bodgea following closely behind. Cluster 3 had only two points in it, with both having the same exact venues and also both are in Staten Island. For cluster 4, Deli/Bodega was in about half the most common venues and in over half of the second most common venue. The final cluster had public locations mostly for its most common and second most common venues. 

For Toronto, the first cluster has mostly public locations for its first and second most common venues. Cluster 2 had two points which both sandwich shops as a first or second most common venue. Cluster 3 had coffee shops as its most most common venue follow by cafe. For the second most common cluster 3 had cafe as its most common venue and coffee shops as it's second most common. Clusters 4 & 5 both had only one point, with music venues being the most common for cluster 4 and cluster 5 had bus line as the most common venue.

Discussion & conclussion:

After looking at the clusters for both cities, they are not similar enough based on this initial analysis. One interesting note is that public locations, like parks, normally fall into being one of the more popular locations. The Toronto dataset does have significantly less datapoints compared to New York and with some the neighborhoods in Toronto being combined into one point could have changed the way that the clusters were developed in Toronto. 

With New York, I had expected to see certain clusters based on some the boroughs. Since New York is a mix of different cultures, these cultures typically have a central hub in part of the city. However, none of the clusters had predominate type of cuisine. This may be due because the dataset for the city was too broad and focusing on a specific borough might have yield different results. 

Ultimately, while a good learning experience about clustering, New York and Toronto are not similar based upon the clustering that I performed in this assignment. If this assignment were to be repeated, I would not group some of the neighborhoods together and just run them through, similar to how New York was done. While the results might be similar, it would be a good check. New York is such a broad city that to ultiamtely compare it to another city, you would probably have to find similar city in size and only use a section of it to compare to one borough at a time.

```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
venues = results['response']['groups'][0]['items']
nearby_venues = json_normalize(venues) # flatten JSON
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]
nearby_venues.head()
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
newyork_venues = getNearbyVenues(names=neighborhoods['Neighborhood'],latitudes=neighborhoods['Latitude'],longitudes=neighborhoods['Longitude'])
```

```python
newyork_onehot = pd.get_dummies(newyork_venues[['Venue Category']], prefix="", prefix_sep="")
newyork_onehot['Neighborhood'] = newyork_venues['Neighborhood'] 
fixed_columns = [newyork_onehot.columns[-1]] + list(newyork_onehot.columns[:-1])
newyork_onehot = newyork_onehot[fixed_columns]
newyork_onehot.head()
```

```python
newyork_grouped = newyork_onehot.groupby('Neighborhood').mean().reset_index()
newyork_grouped
```

```python
num_top_venues = 5

for hood in newyork_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = newyork_grouped[newyork_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 5
indicators = ['st', 'nd', 'rd']
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = newyork_grouped['Neighborhood']
for ind in np.arange(newyork_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(newyork_grouped.iloc[ind, :], num_top_venues)
```

```python
kclusters = 5
newyork_grouped_clustering = newyork_grouped.drop('Neighborhood', 1)
kmeans = KMeans(n_clusters=kclusters,init ='random',random_state=0).fit(newyork_grouped_clustering)
```

```python
neighborhoods_venues_sorted.insert(1, 'Cluster Labels', kmeans.labels_)
newyork_merged = neighborhoods
newyork_merged = newyork_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')
```

```python
newyork_merged = newyork_merged.dropna()
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(newyork_merged['Latitude'], newyork_merged['Longitude'], newyork_merged['Neighborhood'], newyork_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[int(cluster)-1],
        fill=True,
        fill_color=rainbow[int(cluster)-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```

```python
def html_to_pandas(table_html):

    ##header = [s.strip() for s in str(table_html.find_all('th'))[1:-1].replace('<th>','').replace('</th>','').replace('\n','').split(',')]
    
    row_l = [s.find_all(['td','th']) for s in table_html.find_all('tr')]
    
    row_h = []
    row_g = []  
    
    for s in row_l:
        row_g = []
        for c in s:
            row_g.append(c.text.replace('\n',''))
        row_h.append(row_g)
    
    df = pd.DataFrame(row_h)
    
    return df.rename(columns=df.iloc[0]).reindex(df.index.drop(0)).reset_index(drop=True)
    
    
df = html_to_pandas(My_table)
df2 = df
```
